```python
def detect_outlier(val, data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)

    iqr = q3 - q1 

    max_value = q3 + 1.5 * iqr
    min_value = q1 - 1.5 * iqr

    return 1 if (val >= min_value and val <= max_value) else 0 
final['recency_outlier'] = final['recency'].apply(lambda x : detect_outlier(x, final.recency))
final['frequency_outlier'] = final['frequency'].apply(lambda x : detect_outlier(x, final.frequency))
final['monetary_outlier'] = final['monetary'].apply(lambda x : detect_outlier(x, final.monetary))
```

```python
def classifyRecency(val):
    if val <= 21:
            return 4
    elif val <= 80: 
            return 3
    elif val <= 200:
            return 2 
    return 1
```

```python
final['recency_class'] = final.recency.apply(classifyRecency)
```

```python
def classifyFrequency(val):
    if val == 1:
        return 1
    elif val <= 3:
        return 2
    elif val <= 6:
        return 3
    return 4

final['frequency_class'] = final.frequency.apply(classifyFrequency)
```

```python
final
```

```python
final.monetary.describe()
```

```python
def classifyMonetary(val):
    if val<= 180:
        return 1
    elif val<=250:
        return 2
    elif val<=450:
        return 3
    elif val <=750:
        return 4
```

```python
final['monetary_class'] = final.monetary.apply(classifyMonetary)
```

```python
final
```

```python
label_map = {1: "worst", 2: "bad", 3: "good", 4: "best"}
final['Overall'] = (
    'recency: ' + final['recency_class'].map(label_map) +
    ', frequency: ' + final['frequency_class'].map(label_map) +
    ', monetary: ' + final['monetary_class'].map(label_map)
)
```

```python
fig = px.scatter_3d(
    final, 
    x="recency", 
    y="frequency", 
    z="monetary", 
    color="Overall",       
    title="Vizualization"
)
fig.write_html('first_visualization_class.html')
```

```python
score_table = {"frequency" : 10, "recency" : 5, "monetary" : 7}
score_map = {1: 0, 2: 4, 3: 7, 4: 10}



final['score'] = final['frequency_class'].map(score_map) * score_table['frequency'] \
                 + final['recency_class'].map(score_map) * score_table['recency'] \
                 + final['monetary_class'].map(score_map) * score_table['monetary']
```

```python
def scoreLabel(val):
    if val < 60 : 
        return 'worst'
    elif val < 90:
        return 'bad'
    elif val < 120: 
        return 'good'
    return 'best'
```
