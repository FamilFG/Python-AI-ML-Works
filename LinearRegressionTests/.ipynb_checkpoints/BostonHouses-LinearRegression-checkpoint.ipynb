```python
def detect_outlier(val, data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1 
    max_value = q3 + 1.5 * iqr
    min_value = q1 - 1.5 * iqr

    return 1 if (val >= min_value and val <= max_value) else 0
```

```python
cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
for col in cols:
    df[f'{col}_outlier'] = df[col].apply(lambda x: detect_outlier(x, df[col]))
```

```python
x = dfd['RM'].values
y = dfd['MEDV'].values
x_norm = (x - x.mean()) / x.std()

def LinearRegressionwithone(x,y,w=0,b=0, alfa=0.001, n_iterations =10000):
    #cost function
    n = x.shape[0]

    for i in range(n_iterations):
        ypred = w*x + b
        
        #gradients
        dw = 1/n * np.sum((ypred - y)*x)
        db = 1/n * np.sum((ypred - y))
    
        #gradient descents
        w = w - alfa*dw
        b = b - alfa*db
    return w,b

    
w, b = LinearRegressionwithone(x_norm, y)
print("w =", w, "b =", b)

ypred_line = w * x_norm + b
plt.scatter(x, y)
plt.plot(x, ypred_line)
plt.xlabel("RM")
plt.ylabel("MEDV")
plt.show()
```

```python
class MyLinearRegression:
    def __init__(self, alpha=0.01, iterations=100000, normalize=True):
        self.alpha = alpha
        self.iterations = iterations
        self.normalize = normalize
        self.mean_ = None
        self.std_ = None
        self.w = None
        self.b = 0
        #value assignments

    def standardize(self, X):
        self.mean_ = X.mean(axis=0)
        self.std_ = X.std(axis=0)
        self.std_[self.std_ == 0] = 1
        return (X - self.mean_) / self.std_
        #feature scaling

    def transform(self, X):
        return (X - self.mean_) / self.std_
        #transfroming

    def fit(self, X, y):
        if self.normalize:
            X = self.standardize(X)
            
        n, m = X.shape
        self.w = np.zeros(m)
        self.b = 0

        for _ in range(self.iterations):
            pred = X.dot(self.w) + self.b
            difference = pred - y
            dw = (1/n) * X.T.dot(difference)
            db = (1/n) * difference.sum()
            self.w -= self.alpha * dw
            self.b -= self.alpha * db

    def predict(self, X):
        if self.normalize:
            X = self.transform(X)
        return X.dot(self.w) + self.b
```

```python
cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
X = dfd[cols].values
y = dfd['MEDV'].values

model = MyLinearRegression(alpha=0.01, iterations=100000, normalize = True)
model.fit(X, y)
pred = model.predict(X)
```
