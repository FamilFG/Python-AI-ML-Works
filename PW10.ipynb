```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, probplot, norm
```

```python
def load_txt(path):
    return np.loadtxt(path, delimiter=',')
```

```python
def analyze(data, name):
    print(name, "mean =", np.mean(data))
    print(name, "std =", np.std(data))
    print(name, "Shapiro =", shapiro(data))
    plt.hist(data, bins=20)
    plt.title(name)
    plt.show()
    probplot(data, dist="norm", plot=plt)
    plt.title(name)
    plt.show()
def compute_time_delays(timestamps):
    return np.diff(timestamps)
```

```python
X = np.random.normal(1, 2, 1000)
Y = np.random.uniform(-1, 1, 1000)
Z = np.random.exponential(5, 1000)
```

```python
analyze(X, "X")
analyze(Y, "Y")
analyze(Z, "Z")
```

```python
analyze(ph, "pH")
analyze(spinningTop, "spinningTop")
analyze(wind, "wind")
analyze(cosmic, "cosmic")
```
