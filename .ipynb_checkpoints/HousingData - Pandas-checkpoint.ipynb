```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('HousingData.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.dropna(inplace = True)
```

```python
plt.boxplot(df.ZN)
plt.show()
```

```python
cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

for col in cols:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    min_val = q1 - 1.5 * iqr
    max_val = q3 + 1.5 * iqr
    df[f'{col}_outlier'] = (df[col] < min_val) | (df[col] > max_val)

outlier_cols = [col + '_outlier' for col in cols]
rows_with_outliers = df[outlier_cols].any(axis=1)
rows_without_outliers = ~rows_with_outliers
df_clean = df.loc[rows_without_outliers].copy()
df_clean.drop(columns=outlier_cols, inplace=True)
```

```python
df_clean.reset_index(drop=True, inplace=True)
```
