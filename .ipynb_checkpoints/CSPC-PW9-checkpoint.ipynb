```python
import numpy as np
import matplotlib.pyplot as plt

ideal = np.loadtxt("ideal.csv", delimiter=";")
verifideal = np.loadtxt("verifideal.csv", delimiter=";")
noisy = np.loadtxt("noisy.csv", delimiter=";")
print("hello")
```

```python
plt.scatter(ideal[:,0], ideal[:,1], c=ideal[:,2], cmap='bwr', marker='o')
plt.xlabel("v0")
plt.ylabel("alpha")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

X_train, y_train = ideal[:,:2], ideal[:,2]
X_ver, y_ver = verifideal[:,:2], verifideal[:,2]

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn.score(X_train, y_train), knn.score(X_ver, y_ver)
```

```python
confusion_matrix(y_ver, knn.predict(X_ver))
```

```python
from matplotlib.colors import ListedColormap

h = 0.02
x_min, x_max = X_train[:,0].min() - 1, X_train[:,0].max() + 1
y_min, y_max = X_train[:,1].min() - 1, X_train[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.contourf(xx, yy, Z, alpha=0.4, cmap=ListedColormap(['red','blue']))
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=ListedColormap(['red','blue']), edgecolor='k')
plt.xlabel("v0")
plt.ylabel("alfa")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt.score(X_train, y_train), dt.score(X_ver, y_ver)
```

```python
plot_tree(dt)
plt.show()
```

```python
Z = dt.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.contourf(xx, yy, Z, alpha=0.4, cmap=ListedColormap(['red','blue']))
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=ListedColormap(['red','blue']), edgecolor='k')
plt.title("Decision Tree Decision Regions")
plt.xlabel("v0")
plt.ylabel("alfa")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50)
rf.fit(X_train, y_train)
rf.score(X_train, y_train), rf.score(X_ver, y_ver)
```

```python
Z = rf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.contourf(xx, yy, Z, alpha=0.4, cmap=ListedColormap(['red','blue']))
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=ListedColormap(['red','blue']), edgecolor='k')
plt.xlabel("v0")
plt.ylabel("alfa")
plt.show()
```
