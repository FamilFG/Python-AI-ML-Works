```python
class MyLogisticRegression:
    def __init__(self, alfa = 0.1, iterations = 10000, normalize = True):
        self.alfa = alfa
        self.iterations = iterations
        self.normalize = normalize

        self.coef_ = []
        self.weights = []
        self.bias_ = []
        self.cost_function_history = []

    def sigmoid(self, Z):
        return 1/(1+np.exp(-Z))

    def costFunction(self, Y_hat, Y):
        Y_hat = np.clip(Y_hat, 1e-10, 1 - 1e-10)
        return (-Y * np.log(Y_hat) - (1-Y) * np.log(1 - Y_hat)).mean()

    def fit(self, X, Y):
        X_ = np.c_[np.ones(X.shape[0]), X]
        w = np.array([random.randint(1, 10)] * X_.shape[1], dtype=float)

        
        for i in range(self.iterations):
            prediction = self.sigmoid(X_.dot(w))
            cost_result = self.costFunction(prediction, Y)
            self.cost_function_history.append(cost_result)
            diff = Y - prediction
            gradient = -np.dot(X_.T, diff) / X.shape[0]
            w = w -self.alfa * gradient

        self.weights = w.copy()
        self.coef_ = w[1:]
        self.bias_ = [w[0]]

    def predict_proba(self, X):
        pred = self.sigmoid(np.dot(X, self.coef_) + self.bias_)
        return pred

    def predict(self, X):
        probability = self.predict_proba(X)
        probability[probability >=0.5] = 1
        probability[probability <0.5] = 0
        return probability

    
        
```

```python
X = df[['exam_1', 'exam_2']]
y = df['admitted']
```

```python
model = MyLogisticRegression(0.1, 10000)
model.fit(X,y)
print(model.coef_)
```

```python
model2 = LogisticRegression()
model2.fit(X, y)

y_pred = model2.predict(X)
y_prob = model2.predict_proba(X)
print(model2.coef_)
```
