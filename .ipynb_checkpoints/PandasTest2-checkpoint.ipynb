```python
import random
import numpy as np
```

```python
class MyLinearRegression:
    def __init__(self, alpha = 0.1,normalize = True):
        self.learning_rate = alpha
        self.n_iterations = 0
        self.weights = None
        self.normalize = normalize
        self.bias = None
        self.coeficients = None

    def costFunction(X,W,Y):
        pred = X.dot(W)
        difference = (pred-Y)**2
        cost = np.sum(difference)/ (2*X.shape[0])
        return cost

    def standartize(self,X):
        self.mean_ = np.array(X.mean(axis=0))
        self.std_ = np.array(X.std(axis = 0))
        self.std_[self.std_ == 0] =1
        return (X - self.mean_) / self.std_

    def fit(self, X,Y):
        if self.normalize == True:
            X = self.standartize(X)
            
        self.weights = []
        for i in range(X.shape[1]+1):
            self.weights.append(random.randint(1,20))
            
        X_new = np.c_[np.ones(X.shape[0]) , X]
        
        while True:
            pred = X_new.dot(self.weights)
            diff = pred-Y
            gradient = X_new.T.dot(diff)
            if round(gradient[0],3) == 0:
                break
            self.weights = self.weights - self.learning_rate * gradient/X.shape[0]

        self.coeficients = self.weights[1:]
        self.bias = [self.weights[0]]

    def predict(self,X):
        if self.normalize == True:
            X = self.standartize(X)
            
        result = X.dot(self.coeficients + self.bias[0])
        return result
```

```python
X = dfd.drop('Qiymet', axis =1 )
y = dfd['Qiymet']
```

```python
model = MyLinearRegression(normalize = True)
model.fit(X,y)
```
