```python
class Product:
    def __init__(self, pid, pname, pprice):
        self._productId = pid
        self._productName = pname
        self._productPrice = pprice


    def get_productId(self):
        return self._productId

    def get_productName(self):
        return self._productName

    def get_productPrice(self):
        return self._productPrice
        
    def set_productId(self, pid):
        self._productId = pid

    def set_productName(self, pname):
        self._productName = pname

    def set_productPrice(self, pprice):
        self._productPrice = pprice

    def __str__(self):
        return f'Product`s name: {self.get_productName()}\nProduct`s id: {self.get_productId()}\nProduct`s price: {self.get_productPrice()}'
    
        
```

```python
class Inventory:
    def __init__(self):
        self._stock = {}
        
    def __str__(self):
        return f'Our stock: {self.get_stock()}'

    def add_stock(self, product, quantity):
        
        if product.get_productId() not in self._stock:
            self._stock[product.get_productId()] = 0
            
        self._stock[product.get_productId()] += quantity

    def reduce_stock(self, product, quantity):
        if self._stock.get(product.get_productId(), 0) < quantity:
            raise ValueError("Not enough product")
            
        self._stock[product.get_productId()] -= quantity

    def get_stock(self, product):
        return self._stock.get(product.get_productId(), 0)

    
```

```python
class MethodPayment(ABC):
    @abstractmethod
    def pay(self,amount):
        pass
```

```python
class CashPayment(MethodPayment):
    def pay(self, amount):
        print(f'Cash payment done: {amount}')
        
```

```python
class CardPayment(MethodPayment):
    def pay(self, amount):
        print(f'Card payment done: {amount}')
```

```python
class Sale:
    def __init__(self, product, quantity, paymentmethod):
        self.product = product
        self.quantity = quantity
        self.paymentMethod = paymentmethod
        
    def totalPrice(self):
        return self.product.price * self.quantity

    def completePayment(self):
        return self.paymentMethod.pay(self.totalPrice)

    
```
