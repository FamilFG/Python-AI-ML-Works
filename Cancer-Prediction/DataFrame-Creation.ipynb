```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_samples = 200

df = pd.DataFrame({
    "id": np.arange(1, n_samples+1),
    "diagnosis": np.random.choice(["B", "M"], size=n_samples),
    "radius_mean": np.random.normal(14, 3, n_samples),
    "texture_mean": np.random.normal(20, 5, n_samples),
    "perimeter_mean": np.random.normal(90, 20, n_samples),
    "area_mean": np.random.normal(600, 150, n_samples),
    "smoothness_mean": np.random.normal(0.1, 0.02, n_samples),
    "compactness_mean": np.random.normal(0.2, 0.05, n_samples),
    "concavity_mean": np.random.normal(0.15, 0.05, n_samples),
    "concave points_mean": np.random.normal(0.1, 0.03, n_samples),
    "symmetry_mean": np.random.normal(0.2, 0.05, n_samples),
    "fractal_dimension_mean": np.random.normal(0.06, 0.01, n_samples),
    "radius_se": np.random.normal(0.4, 0.1, n_samples),
    "texture_se": np.random.normal(1.0, 0.2, n_samples),
    "perimeter_se": np.random.normal(2.5, 0.5, n_samples),
    "area_se": np.random.normal(40, 10, n_samples),
    "smoothness_se": np.random.normal(0.01, 0.005, n_samples),
    "compactness_se": np.random.normal(0.03, 0.01, n_samples),
    "concavity_se": np.random.normal(0.02, 0.01, n_samples),
    "concave points_se": np.random.normal(0.01, 0.005, n_samples),
    "symmetry_se": np.random.normal(0.02, 0.005, n_samples),
    "fractal_dimension_se": np.random.normal(0.003, 0.001, n_samples),
    "radius_worst": np.random.normal(16, 4, n_samples),
    "texture_worst": np.random.normal(25, 6, n_samples),
    "perimeter_worst": np.random.normal(110, 25, n_samples),
    "area_worst": np.random.normal(800, 200, n_samples),
    "smoothness_worst": np.random.normal(0.12, 0.03, n_samples),
    "compactness_worst": np.random.normal(0.3, 0.07, n_samples),
    "concavity_worst": np.random.normal(0.25, 0.08, n_samples),
    "concave points_worst": np.random.normal(0.15, 0.04, n_samples),
    "symmetry_worst": np.random.normal(0.3, 0.06, n_samples),
    "fractal_dimension_worst": np.random.normal(0.08, 0.02, n_samples)
})

df = pd.get_dummies(df, columns=["diagnosis"], drop_first=True)
df.to_csv("synthetic_cancer_full.csv", index=False)
```
